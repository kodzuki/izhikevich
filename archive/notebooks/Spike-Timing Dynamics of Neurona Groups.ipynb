- #### Paper Spike-Timing Dynamics of Neurona Groups - 29.7.25

----

- División en 2 partes: 

La división es natural porque los aspectos básicos (STDP + delays → formación de grupos) constituyen el hallazgo central, mientras que las extensiones demuestran robustez y propiedades emergentes más complejas.

----

-  #### Parte 1 (Versión Simplificada)

Componentes básicos:

- Modelo neuronal de Izhikevich (ecuaciones 1-2)
- Red de 100,000 neuronas (80,000 excitatorias, 20,000 inhibitorias)
- Conectividad básica y delays de conducción axonal
- STDP básico (regla de potenciación/depresión por timing)
- Dinámica sináptica simple (receptores AMPA, NMDA, GABA)
- Formación espontánea de grupos neuronales
- Caracterización básica (tamaño, distribución espacial, tiempo de duración)

Resultados básicos:

- Comportamiento colectivo (sincronización vs actividad asíncrona)
- Identificación de grupos por timing de spikes
- Distribución de pesos sinápticos
- Crecimiento inicial de grupos
----

- #### Parte 2 - Extensiones Avanzadas:
  
Componentes avanzados:

- Plasticidad sináptica a corto plazo (depresión/facilitación)
- Turnover sináptico (experimentos de robustez)
- Análisis detallado de campos receptivos/proyectivos
- Competencia y cooperación entre grupos
- Análisis de reentrada (reentry)
- Persistencia a largo plazo (simulaciones de 24h)
