```python
import numpy as np
import matplotlib.pyplot as plt

# Configuracin
n_samples = 10000
n_rows, n_cols = 3, 3

# === PANEL 1: GAMMA ===
fig1, axes1 = plt.subplots(n_rows, n_cols, figsize=(12, 10))
fig1.suptitle('Distribucin GAMMA: Variacin de shape y scale', fontsize=14, fontweight='bold')

shapes = np.linspace(0.5, 5, n_rows)
scales = np.linspace(0.2, 1.5, n_cols)

for i, shape in enumerate(shapes):
    for j, scale in enumerate(scales):
        ax = axes1[i, j]
        delays = np.random.gamma(shape, scale, n_samples)
        ax.hist(delays, bins=50, alpha=0.7, edgecolor='black')
        ax.set_xlim(0, 8)
        ax.set_title(f'shape={shape:.1f}, scale={scale:.1f}', fontsize=9)
        ax.axvline(delays.mean(), color='red', linestyle='--', linewidth=1, alpha=0.7)
        if j == 0:
            ax.set_ylabel('Frecuencia', fontsize=8)
        if i == n_rows-1:
            ax.set_xlabel('Delay (ms)', fontsize=8)

plt.tight_layout()
```

```python
# === PANEL 2: LOGNORMAL ===
fig2, axes2 = plt.subplots(n_rows, n_cols, figsize=(12, 10))
fig2.suptitle('Distribucin LOGNORMAL: Variacin de mu y sigma', fontsize=14, fontweight='bold')

mus = np.linspace(-1, 0.5, n_rows)
sigmas = np.linspace(0.3, 1.2, n_cols)

for i, mu in enumerate(mus):
    for j, sigma in enumerate(sigmas):
        ax = axes2[i, j]
        delays = np.random.lognormal(mu, sigma, n_samples)
        ax.hist(delays, bins=50, alpha=0.7, edgecolor='black')
        ax.set_xlim(0, 8)
        ax.set_title(f'mu={mu:.1f}, sigma={sigma:.1f}', fontsize=9)
        ax.axvline(delays.mean(), color='red', linestyle='--', linewidth=1, alpha=0.7)
        if j == 0:
            ax.set_ylabel('Frecuencia', fontsize=8)
        if i == n_rows-1:
            ax.set_xlabel('Delay (ms)', fontsize=8)

plt.tight_layout()
```

```python
# === PANEL 3: WEIBULL ===
fig3, axes3 = plt.subplots(n_rows, n_cols, figsize=(12, 10))
fig3.suptitle('Distribucin WEIBULL: Variacin de shape y scale', fontsize=14, fontweight='bold')

w_shapes = np.linspace(0.5, 3, n_rows)
w_scales = np.linspace(0.5, 2, n_cols)

for i, shape in enumerate(w_shapes):
    for j, scale in enumerate(w_scales):
        ax = axes3[i, j]
        delays = scale * np.random.weibull(shape, n_samples)
        ax.hist(delays, bins=50, alpha=0.7, edgecolor='black')
        ax.set_xlim(0, 8)
        ax.set_title(f'shape={shape:.1f}, scale={scale:.1f}', fontsize=9)
        ax.axvline(delays.mean(), color='red', linestyle='--', linewidth=1, alpha=0.7)
        if j == 0:
            ax.set_ylabel('Frecuencia', fontsize=8)
        if i == n_rows-1:
            ax.set_xlabel('Delay (ms)', fontsize=8)

plt.tight_layout()
plt.show()
```

```python
# === COMPARACIN DIRECTA DE LAS TRES ===
fig4, axes4 = plt.subplots(1, 3, figsize=(15, 4))
fig4.suptitle('Comparacin: Casos tpicos para delays neuronales', fontsize=14, fontweight='bold')

# Gamma: conexiones tpicas
delays_gamma = np.random.gamma(2.5, 0.4, n_samples)
axes4[0].hist(delays_gamma, bins=50, alpha=0.7, edgecolor='black', color='steelblue')
axes4[0].set_title(f'GAMMA (shape=2.5, scale=0.4)\nmedia={delays_gamma.mean():.2f}ms')
axes4[0].set_xlabel('Delay (ms)')
axes4[0].set_xlim(0, 5)

# Lognormal: cola larga
delays_lognorm = np.random.lognormal(-0.2, 0.7, n_samples)
axes4[1].hist(delays_lognorm, bins=50, alpha=0.7, edgecolor='black', color='coral')
axes4[1].set_title(f'LOGNORMAL (mu=-0.2, sigma=0.7)\nmedia={delays_lognorm.mean():.2f}ms')
axes4[1].set_xlabel('Delay (ms)')
axes4[1].set_xlim(0, 5)

# Weibull: decay rpido
delays_weibull = 1.0 * np.random.weibull(1.2, n_samples)
axes4[2].hist(delays_weibull, bins=50, alpha=0.7, edgecolor='black', color='mediumseagreen')
axes4[2].set_title(f'WEIBULL (shape=1.2, scale=1.0)\nmedia={delays_weibull.mean():.2f}ms')
axes4[2].set_xlabel('Delay (ms)')
axes4[2].set_xlim(0, 5)

plt.tight_layout()
plt.show()
```

```python
# === COMPARACIN DIRECTA DE LAS TRES ===
fig4, axes4 = plt.subplots(1, 3, figsize=(15, 4))
fig4.suptitle('Comparacin: Casos tpicos para delays neuronales humanos (0-10ms)', fontsize=14, fontweight='bold')

# Gamma: conexiones tpicas corticales
delays_gamma = np.random.gamma(3.5, 1.0, n_samples)
axes4[0].hist(delays_gamma, bins=50, alpha=0.7, edgecolor='black', color='steelblue')
axes4[0].set_title(f'GAMMA (shape=3.5, scale=1.0)\nmedia={delays_gamma.mean():.2f}ms')
axes4[0].set_xlabel('Delay (ms)')
axes4[0].set_xlim(0, 10)

# Lognormal: cola larga (conexiones distantes)
delays_lognorm = np.random.lognormal(1.0, 0.6, n_samples)
axes4[1].hist(delays_lognorm, bins=50, alpha=0.7, edgecolor='black', color='coral')
axes4[1].set_title(f'LOGNORMAL (mu=1.0, sigma=0.6)\nmedia={delays_lognorm.mean():.2f}ms')
axes4[1].set_xlabel('Delay (ms)')
axes4[1].set_xlim(0, 10)

# Weibull: decay desde conexiones rpidas
delays_weibull = 3.0 * np.random.weibull(1.8, n_samples)
axes4[2].hist(delays_weibull, bins=50, alpha=0.7, edgecolor='black', color='mediumseagreen')
axes4[2].set_title(f'WEIBULL (shape=1.8, scale=3.0)\nmedia={delays_weibull.mean():.2f}ms')
axes4[2].set_xlabel('Delay (ms)')
axes4[2].set_xlim(0, 10)

plt.tight_layout()
plt.show()
```
